```python
import __future__
import matplotlib.pyplot as plt
import urllib,sys
import numpy as np
import xarray as xr

url = 'http://192.168.1.55/thredds/dodsC/testEnhanced/2004050300_eta_211.nc'
dataset = xr.open_dataset(url)
print(dataset.info())

variable = dataset['Z_sfc'][0,:,:]


plt.figure(figsize=(10,7))
ax = plt.axes();
variable.plot.pcolormesh(ax=ax, x='x', y='y', add_colorbar=False, cmap=plt.cm.ocean)
plt.tight_layout()
```
