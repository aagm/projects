````python
payload = {'q': "select * from water_risk_indicators where indicator = 'water_stress' and model= 'bau' and period='year' and type='absolute' order by year asc"}
r = requests.get('https://wri-01.carto.com/api/v2/sql', params=payload)
tableStructure= pd.read_json(json.dumps(r.json()['rows']), orient='records')
tableStructure.head(1)
````

````python

bar = vincent.Bar(tableStructure.value)
bar.axis_titles(x='year', y='Value')
bar.display();
````
